**finding the z-score**

```python
from scipy import stats

# Calculate Z-scores of numerical columns
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))

# Identify outliers (Z-score > 3 or < -3)
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]
```

**remove outliers**

```python
# Remove rows with outliers based on Z-score
df_no_outliers = df[(z_scores < 3).all(axis=1) & (z_scores > -3).all(axis=1)]
df_no_outliers
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line plot of Road Network Length and Public Transport Usage over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Road Network Length (in km)', hue='Country')
plt.title('Road Network Length Over Time')
plt.show()

# Bar plot of Public Transport Usage
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Public Transport Usage (%)')
plt.title('Public Transport Usage by Country')
plt.xticks(rotation=45)
plt.show()
```

**Conclusion :  In this section we conclude that USA use least Public transport and china use most of the public transport**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the data to have years and countries
df_pivot = df.pivot_table(index='Year', columns='Country', values=['Agricultural Land (%)', 'Unemployment Rate (%)', 'Labor Force Participation Rate (%)'])

# Plot heatmap for each feature
plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot['Agricultural Land (%)'], cmap="YlGnBu", annot=False)
plt.title('Agricultural Land (%) Over Time')
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot['Unemployment Rate (%)'], cmap="YlOrRd", annot=False)
plt.title('Unemployment Rate (%) Over Time')
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot['Labor Force Participation Rate (%)'], cmap="PuBu", annot=False)
plt.title('Labor Force Participation Rate (%) Over Time')
plt.show()
```

**Conclusion : In first part The agricultre land over time ranges China is least and India is most nd part will conclude as the unemployment rate in USA and China is most , Canada, Russia and Australia ia greater than average and India is 7.5% ,means most and in third section tells us that The labour focus Prticipant % is less than other country**
